### Features Selection

Based on correlation matrix and plot of correlation, we choose variables which have correlation value with price larger than 0.25. 

```python
all_columns = dataset.columns.tolist()
print(f'All Columns : {all_columns} \n\n')
```

```python
dataset_ = dataset[['price', 'bedrooms', 'bathrooms', 'sqft_living','floors', 'waterfront', 'view','grade', 'sqft_above', 
                    'sqft_basement', 'lat', 'sqft_living15']]
```

```python
dataset_.head()
```

```python
plt.hist(dataset_['price'], bins = 'sqrt', histtype = 'step')
plt.legend(loc = 'upper right')
plt.xlabel('House Price ($USA)')
plt.savefig('Dataset.png', dpi = 300)

plt.show()
```

Based on box plot, there are many outliers. So, we must exclude the outiliers to gain the best model.

```python
dataset_wo = dataset_[(np.abs(stats.zscore(dataset)) < 3).all(axis=1)]
```

```python
dataset_wo
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x='bedrooms', y='price', data=dataset_wo, palette = 'hls').set(xlabel='Number of Bedrooms', ylabel='Price', title='Comparison of House price and number of bedrooms')
plt.savefig('wobbedrooms.png', dpi = 300)
```

```python
plt.hist(dataset_wo['price'], bins = 'sqrt', histtype = 'step')
plt.legend(loc = 'upper right')
plt.xlabel('House Price ($USA)')
plt.savefig('1Dataset.png', dpi = 300)

plt.show()
```

# Perform Feature Standerd Scalling

```python
y = dataset_wo['price']
X = dataset_wo.drop(['price'], axis = 1)
cols = X.columns
index = X.index
```

```python
X.head   
```

```python
# Scale the data to keep all the values in the same magnitude of 0 -1 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X= sc.fit_transform(X)
X = pd.DataFrame(X, columns = cols, index = index)
```

```python
X.head()
```

# Model Implementation

### Train Test Splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=21)
```

```python
X_train
```

```python
plt.hist(y_train, bins = 'sqrt', histtype = 'step', label = 'Data Training')
plt.legend(loc = 'upper right')
plt.xlabel('House Price ($USA)')
plt.savefig('Data_training.png', dpi = 300)

plt.show()
```

```python
X_test
```

```python
plt.hist(y_test, bins = 'sqrt', histtype = 'step', label = 'Data Testing')
plt.legend(loc = 'upper right')
plt.xlabel('House Price ($USA)')
plt.savefig('Data_testing.png', dpi = 300)

plt.show()
```

### Model Selection

#### Import Machine Learning

```python
from sklearn import linear_model

from sklearn.tree import DecisionTreeRegressor

from sklearn.neighbors import KNeighborsRegressor

from sklearn.ensemble import RandomForestRegressor

from sklearn.svm import SVR
```

#### Import Machine Learning Libraries

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

# Decision Tree Regression

### Tunning Parameters

```python
from  sklearn.model_selection import RandomizedSearchCV
```

```python
model = DecisionTreeRegressor()

param_grid = {'max_depth' : np.arange(4, 20), 'min_samples_leaf' : np.arange(3, 20)}
dtc_model = RandomizedSearchCV(model, param_grid, cv = 5)
dtc_model.fit(X_train, y_train)

print("\n Best Score is ")
print(dtc_model.best_score_)

print("\n Best Estimator is ")
print(dtc_model.best_estimator_)

print("\n Best Parametes are")
print(dtc_model.best_params_)
```

```python
y_pred_dtc = dtc_model.predict(X_test)
df_pred_dtc = pd.DataFrame(y_pred_dtc)
```

```python
df_pred_dtc = pd.DataFrame(data=[],columns=["y_pred"])
df_pred_dtc["y_pred"] = y_pred_dtc
df_pred_dtc.index = df_pred_dtc.index
df_pred_dtc.head()
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred_dtc))
print('Root Mean Squared Error of Decision Tree Regression is:', '%.3f'%rmse)
```

```python
r = r2_score(y_test, y_pred_dtc)
print('Coefficient of Determination is:', '%.3f'%r)
```

### Visualization

```python
fig = plt.figure(figsize=(12, 8))
plt.hist(y_pred_dtc, bins = 'sqrt', label = 'Predicted', color = 'green', alpha = 0.5)
plt.hist(y_test, bins = 'sqrt', label = 'Actual', color = 'red', alpha = 0.5)
plt.legend(loc = 'upper right')
plt.xlabel('House Sale Price ($USA)')
plt.title('Decision Tree Regressor')
#plt.xlim(0.01e+6, 6e+6)
plt.savefig('DTC1.png', dpi = 300)

plt.show()
```

# Random Forest Regression

```python
model = RandomForestRegressor()

random_grid = {'bootstrap': [True, False],
 'max_depth': [40, 50, None], # 10, 20, 30, 60, 70, 100,
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2], # , 4
 'min_samples_split': [2, 5], # , 10
 'n_estimators': [200, 400]} # , 600, 800, 1000, 1200, 1400, 1600, 1800, 2000

rfc_model = RandomizedSearchCV(model, random_grid, cv=10) # 10 Cross Validation

rfc_model.fit(X_train, y_train)

print("\n Best Score is ")
print(rfc_model.best_score_) 

print("\n Best Estimator is ")
print(rfc_model.best_estimator_)

print("\n Best Parametes are")
print(rfc_model.best_params_)
```

```python
rfc_model = RandomForestRegressor(n_estimators = 200, min_samples_split = 5,
                                                      min_samples_leaf = 1, max_features = 'sqrt', 
                                                      max_depth = None, bootstrap = True)
rfc_model.fit(X_train, y_train)
```

```python
y_pred_rf = rfc_model.predict(X_test)
df_pred_rf = pd.DataFrame(y_pred_rf)
```

```python
df_pred_rf = pd.DataFrame(data=[],columns=["y_pred"])
df_pred_rf["y_pred"] = y_pred_rf
df_pred_rf.index = df_pred_rf.index
df_pred_rf.head()
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print('Mean Squared Error of Multiple Linear Regression is:', '%.3f'%rmse)
```

```python
fig = plt.figure(figsize=(12, 8))
plt.hist(y_pred_rf, bins = 'sqrt', label = 'Predicted', color = 'green', alpha = 0.5)
plt.hist(y_test, bins = 'sqrt', label = 'Actual', color = 'red', alpha = 0.5)
plt.legend(loc = 'upper right')
plt.xlabel('House Sale Price ($USA)')
plt.title('Random Forest Regressor')
#plt.xlim(0.01e+6, 6e+6)
plt.savefig('RFR.png', dpi = 300)

plt.show()
```

# Support Vector Regression
